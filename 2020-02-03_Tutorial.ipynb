## Functions
Functions, such as `print()` above, need to have parentheses `()` afterward, even if they have no argument. 

We define our own functions with `def`.

With functions (and loops and if statements), we must include a colon `:` after the function statement.

Also, the "inside" of the function is defined by **whitespace**. The level of indentation defines the code that is a part of the function definition. 

You can use any number of tabs or spaces (usually one tab or 2 or 4 spaces), but please, I beg you, please, *never ever use both tabs and spaces*.

`return` outputs things

````python
# First, we define our function
def firstfunction():
    print('Hello, World!')
    
# Then, we run it
firstfunction()


# A function with arguments
def secondfunction(argument):
    return argument**4, argument**3-argument

X, Y = secondfunction(3)
print(X)
print(Y)


# A function with optional arguments
def thirdfunction(arg1, arg2=12):
    return arg1*arg2

print(secondfunction(8))
````

#### Indexing and slicing

You can access objects inside a list, tuple, or even string using square brackets `[]`.

## The Vicsek model

The Vicsek model, developed by Tamás Vicsek in 1994, is the simplest model of collective behavior.

We consider only alignment of neighbors with one another,

$\theta_{i,n+1} = \langle\theta_n\rangle + \eta$

where $\langle\theta_n\rangle$ represents the average angle of neighbors.

````python

















## In case Gary needs a hint ##
import numpy as np
import matplotlib.pyplot as plt
from random import uniform, gauss

num_steps = 1000
num_agents = 20

noise = np.pi/6
r_cutoff = 10

datalist = np.zeros((num_steps, 2*num_agents))

x = np.random.rand(num_agents, 2)
angles = np.zeros((num_agents,1))

for step in range(num_steps):
    next_angles = np.zeros((num_agents,1))
    for agent in range(num_agents):
        close_angles = [angles[agentj] for agentj in range(num_agents) if np.linalg.norm(x[agent,:]-x[agentj,:])<r_cutoff]
        next_angles[agent] = np.mean(close_angles)+noise*uniform(-1, 1)
    angles = next_angles
    for agent in range(num_agents):
        v = np.array([np.cos(angles[agent]), np.sin(angles[agent])])
        x[agent, :] = x[agent, :]+v.reshape((1,2))
    datalist[step, ::2] = x[:,0]
    datalist[step, 1::2] = x[:,1]

plt.figure(figsize=(5,5))
for agent in range(num_agents):
    plt.plot(datalist[:, 2*agent], datalist[:, 2*agent+1])
````

## If time allows: The boids model

There are a couple elements of flocking not included in the Vicsek model that are included in the boids model (boid is a made up word for bird-oid), introduced in a 1987 SIGGRAPH paper by Craig Reynolds, for computer models of flocking.

The boids model includes 3 rules:

1. Collision avoidance
2. Velocity matching
3. Flock centering

We'll keep velocity matching the same as the Vicsek model and implement collision avoidance and flock centering.

To include collision avoidance and flock centering, we will use the Lennard-Jones potential, which models intermolecular forces in gases,

$V_{LJ} = \left(\frac{1}{r^12}-\frac{1}{r^6}\right)$

````python














## In case Gary needs a hint ##
import numpy as np
import matplotlib.pyplot as plt
from random import uniform, gauss

num_steps = 250
num_agents = 20

noise = np.pi/12
r_cutoff = 10

datalist = np.zeros((num_steps, 2*num_agents))

x = np.random.rand(num_agents, 2)
angles = np.zeros((num_agents,1))


def mean_angles(x, angles):
    next_angles = np.zeros((num_agents,1))
    for agent in range(num_agents):
        close_angles = [angles[agentj] for agentj in range(num_agents) 
                        if np.linalg.norm(x[agent,:]-x[agentj,:])<r_cutoff]
        next_angles[agent] = np.mean(close_angles)
    return next_angles

def lennard_jones(x, angles):
    v = np.zeros((num_agents,2))
    for agent in range(num_agents):
        close_positions = np.array([x[agentj,:] for agentj in range(num_agents) 
                           if np.linalg.norm(x[agent,:]-x[agentj,:])<r_cutoff
                           and not agent==agentj])
        for neighbor_n in range(len(close_positions)):
            r = close_positions[neighbor_n, :]-x[agent, :]
            if np.linalg.norm(r) > 0:
               v[agent, :] += 48/r**2*(1/r**12-1/r**6)
    return v

def random_angle(noise):
    return noise*uniform(-1, 1)

for step in range(num_steps):
    next_angles = mean_angles(x, angles)
    LJforce = lennard_jones(x, angles)
    for agent in range(num_agents):
        LJangle = np.arctan2(LJforce[agent, 1], LJforce[agent, 0])
        angles[agent] = LJangle+next_angles[agent]+random_angle(noise)
        v = np.array([np.cos(angles[agent]), np.sin(angles[agent])]).reshape((1,2))
        x[agent, :] = x[agent, :]+v
    datalist[step, ::2] = x[:,0]
    datalist[step, 1::2] = x[:,1]

for agent in range(num_agents):
    plt.plot(datalist[:, 2*agent], datalist[:, 2*agent+1])

print(datalist[-1, :])
````
